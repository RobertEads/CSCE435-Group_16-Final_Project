```python
import sys
sys.path.append("/scratch/group/csce435-f23/python-3.8.17/lib/python3.8/site-packages")
sys.path.append("/scratch/group/csce435-f23/thicket")
from glob import glob
import matplotlib.pyplot as plt
import pandas as pd
import thicket as th

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
```

```python
tk = th.Thicket.from_caliperreader(glob("*s16*.cali"))
```

```python
tk.dataframe
```

```python
tk.metadata
```

```python
tk.statsframe.dataframe["time"] = 1
```

```python
print(tk.tree())
```

```python
gb = tk.groupby("InputType")
```

```python
ctk = th.Thicket.concat_thickets(
    thickets=list(gb.values()),
    headers=list(gb.keys()),
    axis='columns',
    metadata_key='num_threads'
)
```

```python
ctk.dataframe
```

```python
ctk.dataframe = ctk.dataframe.reset_index().drop([
                                                  ("node", ""), ('Randomized', "nid"), ('Randomized', "spot.channel"),('Randomized', "Min time/rank"), ('Randomized', "Max time/rank"),('Randomized', "Total time"), ('Randomized', "Avg GPU time/rank"), ('Randomized', "Min GPU time/rank"),('Randomized', "Total GPU time"), ('Randomized', "Max GPU time/rank"),
                                    
                                                  ("node", ""), ('Sorted', "nid"), ('Sorted', "spot.channel"),('Sorted', "Min time/rank"), ('Sorted', "Max time/rank"),('Sorted', "Total time"), ('Sorted', "Avg GPU time/rank"), ('Sorted', "Min GPU time/rank"),('Sorted', "Total GPU time"), ('Sorted', "Max GPU time/rank")
                                                 ], axis=1)
```

```python
ctk.dataframe = ctk.dataframe.rename({("name", ""): "name", ("num_threads", ""): "num_threads"}).set_index(["name", "num_threads"])
ctk.dataframe
```

```python
main = ctk.dataframe.loc["main"]
main
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
legend_labels = []
for column in main.columns:
    first_index = column[0]  # Extract the first index
    legend_labels.append(first_index)
    plt.plot(main.index, main.xs(column, axis=1), marker='o', label=column)

plt.xlabel('Number of Threads')
plt.ylabel('Avg time per rank (s)')
plt.title('CUDA - Weak Scaling for 2^16 Elements')
plt.legend(legend_labels)
plt.grid(True)
plt.show()
```

```python
# Drop unwanted columns
ctk.dataframe = ctk.dataframe.reset_index().drop([("node", ""), (65536, "nid"), (1048576, "nid"), (262144, "nid"),(65536, "spot.channel"), (1048576, "spot.channel"), (262144, "spot.channel"), (65536, "Avg time/rank"), (1048576, "Avg time/rank"), (262144, "Avg time/rank")], axis=1)
ctk.dataframe = ctk.dataframe.reset_index().drop([(65536, "Min time/rank"), (1048576, "Min time/rank"), (262144, "Min time/rank"),(65536, "Max time/rank"), (1048576, "Max time/rank"), (262144, "Max time/rank"),(65536, "Variance time/rank"), (1048576, "Variance time/rank"), (262144, "Variance time/rank")], axis=1)
# Rename index to remove tuple
ctk.dataframe = ctk.dataframe.rename({("name", ""): "name", ("num_procs", ""): "num_procs"}).set_index(["name", "num_procs"])
```

```python
ctk.dataframe
```

```python
# Define common variables
function_names = ["main","comm","comp","data_init","correctness_check "]
threads = sorted(list(set(ctk.dataframe.index.get_level_values("num_procs"))))
sizes = [65536, 262144, 1048576]
dataType = "Reverse"
```

```python
def plot_mpl(df, df2, df3, title, xlabel, ylabel, columns):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(
        df,
        marker='o',
    )
    ax.plot(
        df2,
        marker='o',
    )
    ax.plot(
        df3,
        marker='o',
    )
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(columns)
    plt.grid(True)
    plt.yscale("log", base=10)
    plt.xscale("log", base=2)
    plt.xticks(threads)
    plt.show()
    return fig, ax
```

```python
for name in function_names:
    plot_mpl(
            df=ctk.dataframe.loc[name][65536]["Total time"],
            df2=ctk.dataframe.loc[name][262144]["Total time"],
            df3=ctk.dataframe.loc[name][1048576]["Total time"],
            title="Total time-" + name + "-" + dataType + "-Strong_Scaling",
            xlabel="Number of Processes",
            ylabel="Time (seconds)",
            columns=sizes
        )
```

```python
# Drop unwanted columns
ctk.dataframe = ctk.dataframe.reset_index().drop([("node", ""), (65536, "nid"), (1048576, "nid"), (262144, "nid"),(65536, "spot.channel"), (1048576, "spot.channel"), (262144, "spot.channel")], axis=1)
ctk.dataframe = ctk.dataframe.reset_index().drop([(65536, "Min time/rank"), (1048576, "Min time/rank"), (262144, "Min time/rank"),(65536, "Max time/rank"), (1048576, "Max time/rank"), (262144, "Max time/rank"),(65536, "Variance time/rank"), (1048576, "Variance time/rank"), (262144, "Variance time/rank"), (65536, "Avg time/rank"), (1048576, "Avg time/rank"), (262144, "Avg time/rank")], axis=1)
# Rename index to remove tuple
ctk.dataframe = ctk.dataframe.rename({("name", ""): "name", ("num_procs", ""): "num_procs"}).set_index(["name", "num_procs"])
```

```python
ctk.dataframe
```

```python
# Define common variables
numProcs = [16,64,256]
dataType = "Randomized"
functions = ["main", "comm", "comp"]
```

```python
def plot_mpl(df, title, xlabel, ylabel, columns):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(
        df,
        marker='o',
    )
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(columns)
    plt.grid(True)
    plt.yscale("log", base=10)
    plt.xscale("log", base=2)
    plt.xticks(numProcs)
    plt.show()
    return fig, ax
```

```python
for func in functions:
    tempData = [ctk.dataframe.loc[func][65536].at[16, "Total time"], ctk.dataframe.loc[func][262144].at[64, "Total time"], ctk.dataframe.loc[func][1048576].at[256, "Total time"]]
    dfTemp = pd.DataFrame(tempData, columns=["Total time"], index=numProcs) 
    plot_mpl(
                df=dfTemp,
                title= func + "-" + dataType + "-Weak_Scaling",
                xlabel="Number of Processes",
                ylabel="Time (seconds)",
                columns=["Total time"]
            )
```

# Strong Scaling for 2^18 Elements

```python
tk = th.Thicket.from_caliperreader(glob("*n262144*.cali"))
tk.statsframe.dataframe["time"] = 1
gb = tk.groupby("InputType")
ctk = th.Thicket.concat_thickets(
    thickets=list(gb.values()),
    headers=list(gb.keys()),
    axis='columns',
    metadata_key='num_threads'
)
ctk.dataframe = ctk.dataframe.reset_index().drop([("node", ""), ('perturbed_array', "nid"), ('perturbed_array', "spot.channel"),('perturbed_array', "Min time/rank"), ('perturbed_array', "Max time/rank"),('perturbed_array', "Total time"), ('perturbed_array', "Avg GPU time/rank"), ('perturbed_array', "Min GPU time/rank"),('perturbed_array', "Total GPU time"), ('perturbed_array', "Max GPU time/rank"),
                                                  ("node", ""), ('random_array', "nid"), ('random_array', "spot.channel"),('random_array', "Min time/rank"), ('random_array', "Max time/rank"),('random_array', "Total time"), ('random_array', "Avg GPU time/rank"), ('random_array', "Min GPU time/rank"),('random_array', "Total GPU time"), ('random_array', "Max GPU time/rank"),
                                                  ("node", ""), ('reversed_array', "nid"), ('reversed_array', "spot.channel"),('reversed_array', "Min time/rank"), ('reversed_array', "Max time/rank"),('reversed_array', "Total time"), ('reversed_array', "Avg GPU time/rank"), ('reversed_array', "Min GPU time/rank"),('reversed_array', "Total GPU time"), ('reversed_array', "Max GPU time/rank"),
                                                  ("node", ""), ('sorted_array', "nid"), ('sorted_array', "spot.channel"),('sorted_array', "Min time/rank"), ('sorted_array', "Max time/rank"),('sorted_array', "Total time"), ('sorted_array', "Avg GPU time/rank"), ('sorted_array', "Min GPU time/rank"),('sorted_array', "Total GPU time"), ('sorted_array', "Max GPU time/rank")
                                                 ], axis=1)

ctk.dataframe = ctk.dataframe.rename({("name", ""): "name", ("num_threads", ""): "num_threads"}).set_index(["name", "num_threads"])

main_function = ctk.dataframe.loc["main_function"]
comm = ctk.dataframe.loc["comm"]
comm_large = ctk.dataframe.loc["comm_large"]
cudaMemcpy = ctk.dataframe.loc["cudaMemcpy"]
comp = ctk.dataframe.loc["comp"]
comp_large = ctk.dataframe.loc["comp_large"]
correctness_check = ctk.dataframe.loc["correctness_check"]
data_init = ctk.dataframe.loc["data_init"]

regions = [main_function, comm, comm_large, cudaMemcpy, comp, comp_large, correctness_check, data_init]
names = ["main_function", "comm", "comm_large", "cudaMemcpy", "comp", "comp_large", "correctness_check", "data_init"]

for region, name in zip(regions, names):
    plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
    legend_labels = []
    for column in region.columns:
        first_index = column[0]  # Extract the first index
        legend_labels.append(first_index)
        plt.plot(region.index, region.xs(column, axis=1), marker='o', label=column)

    plt.xlabel('Number of Threads')
    plt.ylabel('Avg time per rank (s)')
    plt.title(f'CUDA Merge Sort - Strong Scaling for 2^18 Elements: {name}')
    plt.legend(legend_labels)
    plt.grid(True)
    plt.show()
```

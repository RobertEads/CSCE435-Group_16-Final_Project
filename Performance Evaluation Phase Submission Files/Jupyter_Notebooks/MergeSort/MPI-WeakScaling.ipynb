```python
# WEAK SCALING FOR 2^16
tk = th.Thicket.from_caliperreader(glob("*n65536*.cali"))
```

```python
tk.metadata
```

```python
tk.statsframe.dataframe["time"] = 1
```

```python
print(tk.tree())
```

```python
gb = tk.groupby("InputType")
```

```python
ctk = th.Thicket.concat_thickets(
    thickets=list(gb.values()),
    headers=list(gb.keys()),
    axis='columns',
    metadata_key='num_procs'
)
```

```python
ctk.dataframe
```

```python
ctk.dataframe = ctk.dataframe.reset_index().drop([("node", ""), ('1% Perturbed', "nid"), ('1% Perturbed', "spot.channel"),('1% Perturbed', "Min time/rank"), ('1% Perturbed', "Max time/rank"),('1% Perturbed', "Total time"), ('1% Perturbed', "Variance time/rank"),
                                                 ('Randomized', "nid"), ('Randomized', "spot.channel"),('Randomized', "Min time/rank"), ('Randomized', "Max time/rank"),('Randomized', "Total time"), ('Randomized', "Variance time/rank"),
                                                 ('Reverse Sorted', "nid"), ('Reverse Sorted', "spot.channel"),('Reverse Sorted', "Min time/rank"), ('Reverse Sorted', "Max time/rank"),('Reverse Sorted', "Total time"), ('Reverse Sorted', "Variance time/rank"),
                                                 ('Sorted', "nid"), ('Sorted', "spot.channel"),('Sorted', "Min time/rank"), ('Sorted', "Max time/rank"),('Sorted', "Total time"), ('Sorted', "Variance time/rank"),
                                                 
                                                 ], axis=1)


```

```python
ctk.dataframe = ctk.dataframe.rename({("name", ""): "name", ("num_procs", ""): "num_procs"}).set_index(["name", "num_procs"])
ctk.dataframe
```

```python
main = ctk.dataframe.loc["main_function"]
main
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
legend_labels = []
for column in main.columns:
    first_index = column[0]  # Extract the first index
    legend_labels.append(first_index)
    plt.plot(main.index, main.xs(column, axis=1), marker='o', label=column)

plt.xlabel('Number of Processors')
plt.ylabel('Avg time per rank (s)')
plt.title('Weak Scaling for 2^16 Elements')
plt.legend(legend_labels)
plt.grid(True)
plt.show()
```

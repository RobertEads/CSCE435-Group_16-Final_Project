```python
# Project using Thicket
import sys
sys.path.append("/scratch/group/csce435-f23/python-3.8.17/lib/python3.8/site-packages")
sys.path.append("/scratch/group/csce435-f23/thicket")
from glob import glob

import matplotlib.pyplot as plt
import pandas as pd

import thicket as th

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
```

```python
#Read in all the files
tk = th.Thicket.from_caliperreader(glob("*.cali"))
```

```python
tk.statsframe.dataframe["time"] = 1
print(tk.tree())
```

```python
# Performance data
tk.dataframe
```

```python
#Group by matrix size argument, available in the metadata (tk.metadata) to organize our performance data into 3 groupings.
tk.metadata
```

```python
gb = tk.groupby(["InputType", "InputSize"])
```

```python
#Compose these 3 groups column-wise to get a holistic view of the data
ctk = th.Thicket.concat_thickets(
    thickets=list(gb.values()),
    headers=list(gb.keys()),
    axis="columns",
    metadata_key="num_threads"
)
```

```python
ctk.dataframe
```

```python
#Perform pandas operations to get the table in a desirable format for plotting

# Drop unwanted columns
ctk.dataframe = ctk.dataframe.reset_index().drop([("node", ""), ("Randomized", "nid"), ("Randomized", "spot.channel"), ("Randomized", "Min time/rank"), ("Randomized", "Max time/rank"), ("Randomized", "Total time"), ("Randomized", "Avg GPU time/rank"), ("Randomized", "Min GPU time/rank"), ("Randomized", "Max GPU time/rank"), ("Randomized", "Total GPU time"), ("Reverse Sorted", "nid"), ("Reverse Sorted", "spot.channel"), ("Reverse Sorted", "Min time/rank"), ("Reverse Sorted", "Max time/rank"), ("Reverse Sorted", "Total time"), ("Reverse Sorted", "Avg GPU time/rank"), ("Reverse Sorted", "Min GPU time/rank"), ("Reverse Sorted", "Max GPU time/rank"), ("Reverse Sorted", "Total GPU time"), ("Sorted", "nid"), ("Sorted", "spot.channel"), ("Sorted", "Min time/rank"), ("Sorted", "Max time/rank"), ("Sorted", "Total time"), ("Sorted", "Avg GPU time/rank"), ("Sorted", "Min GPU time/rank"), ("Sorted", "Max GPU time/rank"), ("Sorted", "Total GPU time")], axis=1)

# Rename index to remove tuple
ctk.dataframe = ctk.dataframe.rename({("name", ""): "name", ("num_threads", ""): "num_threads"}).set_index(["name", "num_threads"])
```

```python
ctk.dataframe
```

```python
# Define common variables
function_names = set(ctk.dataframe.index.get_level_values("name"))
procedures = sorted(list(set(ctk.dataframe.index.get_level_values("num_threads"))))
matrix_sizes = sorted(ctk.dataframe.columns.tolist())
```

```python
# Plotting
#Use `matplotlib.pyplot` directly for more fine-grained control. Specifically for our case we can change the x axis scaling from base 10 to base 2.
def plot_mpl(df, title, xlabel, ylabel, keys):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(
        df,
        marker='o',
    )
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(keys)
    plt.grid(True)
    plt.yscale("log", base=10)
    plt.xticks(procedures)
    plt.show()
    return fig, ax
```

```python
main = ctk.dataframe.loc["main"]
main
```

```python
plt.figure(figsize=(10, 6))
legend_labels = []

for column in main.columns:
    first_index = column[0]
    legend_labels.append(first_index)
    plt.plot(main.index, main.xs(column, axis=1), marker='o', label=column)
    
plt.title('Bubble Sort 2^16 Inputs - Main Function - CUDA')
plt.xlabel('Number of Threads')
plt.ylabel('Time (seconds)')
plt.legend(legend_labels)
plt.grid(True)
plt.show()
```
